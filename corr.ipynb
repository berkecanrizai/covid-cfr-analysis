```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math

from scipy.stats import norm
from scipy.stats import chi2

import scipy.stats as sc 
import statsmodels.graphics.gofplots as sm
import os

data=pd.read_excel('InputCFREstimates.xlsx', engine='openpyxl')

#print(data)
data.describe()

data = {'CFR': data['CFR'],
        'Pop Density': data['Density'],
        'GDP Per Capita': data['GDP'],
        'Number of Hospitals': data['Number of Hospitals'],
        'Beds per thousand people': data['Beds per thousand people'],
        'Birth rate': data['Birth rate'],
        'Life expectancy': data['Life expectancy'],
        'Debt per Capita in US dollars': data['Debt per Capita in US dollars'],
        'Air Pollution (micro grams per cubic meter)': data['Air Pollution (micro grams per cubic meter)'],
        'FCFR': data['FCFR'],
        'CVD': data['CVD']
        }

df = pd.DataFrame(data,columns=['CFR','Pop Density','GDP Per Capita','Number of Hospitals'])
corrMatrix = df.corr()
print (corrMatrix)

```
